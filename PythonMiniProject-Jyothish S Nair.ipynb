```python
library = {
               "The Odyssey":{"Author":"Homer","Available":True,"IssueTo":None,"IssueDate":None,"ReturnDate":None},
               "The Great Gatsbay":{"Author":"Scott","Available":True,"IssueTo":None,"IssueDate":None,"ReturnDate":None}
}
```

```python
def welcome():
    print("*"*50)
    print("     Welcome to the Library Management System     ")
    print("*"*50)
    print("Your gateway to efficient book management,\nstudent records, and seamless library services.")
    print()
    print("Please select an option from the menu to continue.\n")
```

```python
#ViewBook
def viewBooks():
    import pandas as pd
    print("-"*50)
    print()
    df = pd.DataFrame([{"BookTitle": k, "Author": v["Author"], "Available": v["Available"],
                        "IssueTo":v["IssueTo"],"IssueDate":v["IssueDate"],"ReturnDate":v["ReturnDate"]}
                        for k, v in library.items()])
    df.index = range(1, len(df) + 1)
    print(df)
```

```python
#AddBooks
def AddNewBook():
    while True:
        
        BookTitle = input("Enter the book title: ")
        Author = input("Enter the author: ")
    
        # Check if book already exists (ignoring case)
        if any(title.lower() == BookTitle.lower() for title in library.keys()):
            print(f"The book '{BookTitle}' by {Author} already exists in the system!")
        else:
            library[BookTitle] = {
                "Author": Author,
                "Available": True,
                "IssueTo": None,
                "IssueDate": None,
                "ReturnDate": None
            }
            print("Added New Book in the Library Management System Successfully")
    
        # Ask if user wants to add another book
        ch = input("Do you want to add another book in the system (Yes/No): ")
        if ch.lower() == "no":
            print("Back to the Library Management System main menu.")
            break
```

```python
#Issue a Book
def IssueBook():
    from datetime import date
    
    while True:
        BookTitle = input("Enter the book title: ")
        IssueTo = input("Enter your Name: ")
        
        # Find the actual key ignoring case
        matched_key = None
        for title in library.keys():
            if title.lower() == BookTitle.lower():
                matched_key = title
                break
        
        if matched_key:  # Book exists
            if library[matched_key]["Available"] == True:
                library[matched_key].setdefault("IssueTo", [])
                library[matched_key]["IssueTo"].append(IssueTo)
                library[matched_key].setdefault("IssueDate", [])
                library[matched_key]["IssueTo"].append(date.today())
                library[matched_key].update({
                    "Available": False,
                    "IssueTo": IssueTo,
                    "IssueDate": date.today(),
                    "ReturnDate": None
                })
                print(f"The book '{matched_key}' issued to {IssueTo} on {date.today()} and updated in the system.")
            else:
                print(f"The book '{matched_key}' is currently not available.")
        else:
            print(f"The book '{BookTitle}' does not exist in the library system.")
        
        # Ask if user wants to check another book
        ch = input("Do you want to check another book in the system (Yes/No): ")
        if ch.lower() == "no":
            print("Back to the Library Management System main menu.")
            break
```

```python
#View Issued Books
def viewIssueBooks():
    import pandas as pd
    print("-"*50)
    print()
    df = pd.DataFrame([{"BookTitle": k, "Author": v["Author"], "Available": v["Available"],
                        "IssueTo":v["IssueTo"],"IssueDate":v["IssueDate"],"ReturnDate":v["ReturnDate"]}
                        for k, v in library.items()])
    df.index = range(1, len(df) + 1)
    IssueBook = df[df['Available']==False]
    print(IssueBook)
```

```python
#Return a Book
def ReturnBooks():
    from datetime import date
    
    while True:
        BookTitle = input("Enter the book title: ")
        
        # Find the actual key ignoring case
        matched_key = None
        for title in library.keys():
            if title.lower() == BookTitle.lower():
                matched_key = title
                break
        
        if matched_key:  # Book exists
            if library[matched_key]["Available"] == False:
                library[matched_key].setdefault("ReturnDateHistory", [])
                library[matched_key]["ReturnDateHistory"].append(date.today())
                library[matched_key].update({
                    "Available": True,
                    "ReturnDate": date.today()
                })
                
                # If you want to keep a history of return dates:

                
                print(f"The book '{matched_key}' returned on {date.today()} and updated in the system.")
            else:
                print(f"The book '{matched_key}' is already returned.")
        else:
            print(f"The book '{BookTitle}' does not exist in the library system.")
        
        # Ask if user wants to check another book
        ch = input("Do you want to return another book in the system (Yes/No): ")
        if ch.lower() == "no":
            print("Back to the Library Management System main menu.")
            break
```

```python
def Menu():
    welcome()
    print("-"*50)
    print("1) View Books")
    print("2) Add a Book")
    print("3) Issue a Book")
    print("4) Return a Book")
    print("5) View Issued Books")
    print("6) Exit")
    print("-"*50)
```

```python
###### Main Program
while True:
    try:
        Menu()
        ch = int(input("Enter your choice: "))

        match ch:
            case 1:
                print("List of Available books are:\n")
                viewBooks()
            case 2:
                AddNewBook()
            case 3:
                
                IssueBook()
            case 4:
                ReturnBooks()
            case 5:
                viewIssueBooks()
            case 6:
                print("Library Management System - you have exited successfully")
                break
            case _:
                print("\tYou have entered a wrong operation - try again")

    except ValueError:
        print("\tDon't enter strings or symbols for choice - try again")
```
