Q1. Create the first numpy array of 1-D with values (1,2,3,0). 
Create a second numpy array of 1-D with values (4,1,0, 
6). Perform following simple arithmetic operations on 
these two numpy arrays: 
• Addition 
• Subtraction 
• Multiplication 
• Division 

```python
import numpy as np
import warnings
warnings.filterwarnings("ignore")
arr1 = np.array([1,2,3,0])
print("arr1 = ",arr1,type(arr1))
print("Shape of arr1 = ",arr1.shape)
print("Dimension of arr1 = ",arr1.ndim)
print("*"*40)
arr2 =np.array([4,1,0,6])
print("arr2 = ",arr2,type(arr2))
print("Shape of arr2 = ",arr1.shape)
print("Dimension of arr2 = ",arr1.ndim)
print("*"*40)
print("Addition of two numpy arrays")
print("-"*30)
Add = arr1+arr2
print("Additon :",Add,type(Add))
print("Shape of Additon Output = ",Add.shape)
print("Dimension of Additon Output = ",Add.ndim)
print("*"*40)
print("Subtraction of two numpy arrays")
print("-"*30)
Sub = arr1-arr2
print("Subtraction :",Sub,type(Sub))
print("Shape of Subtraction Output = ",Sub.shape)
print("Dimension of Subtraction Output = ",Sub.ndim)
print("*"*40)
print("Multiplication of two numpy arrays")
print("-"*30)
Mul = arr1*arr2
print("Multiplication :",Mul,type(Mul))
print("Shape of Multiplication Output = ",Mul.shape)
print("Dimension of Multiplication Output = ",Mul.ndim)
print("*"*40)
print("Division of two numpy arrays")
print("-"*30)
Div = np.divide(arr1, arr2, out=np.zeros_like(arr1, dtype=float), where=arr2!=0)
print("Division :",Div,type(Div))
print("Shape of Division Output = ",Div.shape)
print("Dimension of Division Output = ",Div.ndim)
print("*"*40)
```

Q4 ) Create a 2-D numpy array of random integers between 
1 to 100 with shape 5*4. Now perform following 
operations: 
• Print only the even numbers from this array 
• Print only the odd numbers from this array 
• Print row wise sum of all elements 
• Print column wise sum of all elements 
• Convert this 2-D array into 3-D array of shape 
2*2*5

```python
import numpy as np
arr = np.random.randint(1,100,20).reshape(5,4)
print("arr:")
print(arr)
print("*"*50)
print("Dimesion of arr {}-D".format(arr.ndim))
print("*"*50)
```

```python
def iseven(n):
    return n%2==0
```

```python
print("1) Print only the even numbers from this array")
print("Output:",arr[np.where(iseven(arr))])
print("-"*40)
print("2) Print only the odd numbers from this array")
print("Output:",arr[np.where(~iseven(arr))])
print("-"*40)
print("3) Print row wise sum of all elements")
print("Output:",arr.sum(axis=0))
print("-"*40)
print("4) Print column wise sum of all elements")
print("Output:",arr.sum(axis=1))
print("-"*40)
print("5) Convert this 2-D array into 3-D array of shape 2*2*5")
print("Output:")
print(arr.reshape(2,2,5))
print("Dimension of Output {}-D".format(arr.reshape(2,2,5).ndim))
print("-"*40)
```

Q6. Explain the concept of stacking in numpy with examples. 
How is it different from concatenation. Give any 3 
differences between them. 

##### Example of Stacking

```python
import numpy as np

stack1 = np.arange(1,20,4)
stack2 = np.random.randint(1,50,5)
print("stack1: {} Type: {} Dimension: {}-D".format(stack1,type(stack1),stack1.ndim))
print("stack2: {} Type: {} Dimension: {}-D".format(stack2,type(stack2),stack2.ndim))
print("*"*70)
print("*"*70)
Stacking_row = np.stack((stack1,stack2))
print("Stacking_row:")
print(Stacking_row)
print("Type: {} Dimension: {}-D Shape:{}".format(type(Stacking_row),Stacking_row.ndim,Stacking_row.shape))
print("*"*70)
print("*"*70)
Stacking_col = np.stack((stack1,stack2),axis=1)
print("Stacking_row:")
print(Stacking_col)
print("Type: {} Dimension: {}-D Shape:{}".format(type(Stacking_col),Stacking_col.ndim,Stacking_col.shape))
```

```python
import numpy as np

stack1 = np.arange(1,20,4)
stack2 = np.random.randint(1,50,5)
print("stack1: {} Type: {} Dimension: {}-D".format(stack1,type(stack1),stack1.ndim))
print("stack2: {} Type: {} Dimension: {}-D".format(stack2,type(stack2),stack2.ndim))
print("*"*70)
print("*"*70)
Concat = np.concatenate((stack1,stack2))
print("Concat:",Concat)
print("Type: {} Dimension: {}-D Shape:{}".format(type(Concat),Concat.ndim,Concat.shape))
```
